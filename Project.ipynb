```
SF = SF[(SF['Source Port']!='')&(SF['Destination Port']!='')]
len(SF)
```

```
iplist = ['147.32.84.165','147.32.84.191','147.32.84.192','147.32.84.193','147.32.84.204','147.32.84.205','147.32.84.206','147.32.84.207','147.32.84.208','147.32.84.209']

iplist = turicreate.SArray(iplist)
iplist = iplist.unique() 
iplist = list(iplist)

def fcheckIP(x):
    if (x['Source'] in iplist) or (x['Destination'] in iplist):
        return 1
    else:
        return 0
        

SF['isBot'] = SF[['Source','Destination']].apply(lambda x: fcheckIP(x))
SF.head()
```

```
def flow_id(x):
    
    if x['Source']>x['Destination']:
       
        return x['Source']+'-'+x['Destination']+'-'+str(x['Source Port'])+'-'+str(x['Destination Port'])+'-'+x['Protocol']
    else:
       
        return x['Destination']+'-'+x['Source']+'-'+str(x['Destination Port'])+'-'+str(x['Source Port'])+'-'+x['Protocol']
SF['Flow name'] = SF.apply(lambda x:flow_id(x))
```

```
SF.head(15)
```

```
## function for comparing two different flows based on columns
def compareUF(x,y):
    if x!=y:
        return False
    return True
```

```
#For identifying IOPR
SF['Forward'] = SF.apply(lambda x: 1 if x['Source']>x['Destination'] else 0 )
```

```
SF.head()
```

```
#ฐาน16
SF['tcp_Flags'] = SF['tcp_Flags'].apply(lambda x:int(x,16) if x!='' else 0)
```

```
SF.head()
```

```
##Code for logic of Bidirectional flow identification
import pickle 

FlowNo = 0 ##Unique Flow Number for each flow
prev = None
Flow = []     ##Stores all flows in form of list of dictionary 
#cFlow = []    ##Store the current flow (all details)
count = 0
# fc = 0
startTime = None   ##Start Time of each flow to implement timeout
#SF = SF.sort(['UFid','Time'])
# print 'Done Sorting'
for x in SF:
    if count%500000 == 0:
        print('Running'+str(count)+' Done !')
        
    count = count+1
    
    if prev is None:
        if startTime is None:
            startTime = x['Time']
        Flow.append(FlowNo)
        
        
        prev = x['Flow name']
    elif compareUF(x['Flow name'],prev):
        if x['tcp_Flags']&1 != 0:
            Flow.append(FlowNo)
            prev = None
            startTime = None
            FlowNo = FlowNo + 1
        elif x['tcp_Flags']&4 != 0:
            FlowNo = FlowNo + 1
            Flow.append(FlowNo)
            prev = x['Flow name']
            startTime = x['Time']


        elif x['Time']-startTime>=60:
            FlowNo = FlowNo + 1
            Flow.append(FlowNo)
            prev = None
            startTime = x['Time']
            
            
        else:
            
            Flow.append(FlowNo)
            prev = x['Flow name']
            

    else:
        FlowNo = FlowNo + 1
        Flow.append(FlowNo)
        prev = x['Flow name']
        startTime = x['Time']
```

```
## Ratio of incoming to outgoing packets
temp = SF.groupby('Flow id',{
        'NumForward' : turicreate.aggregate.SUM('Forward'),
        'Total' : turicreate.aggregate.COUNT()
    })
temp['IOPR']= temp.apply(lambda x: ((x['Total']-x['NumForward'])*1.0)/x['NumForward'] if x['NumForward'] !=0 else (-1) )
SF = SF.join(temp,on='Flow id')
SF.head()
```

```
## First Packet Length
FPL = SF.groupby(['Flow id'],{
        'Time':turicreate.aggregate.MIN('Time')
    })
FPL.head(10)
```

```
FPL = FPL.join(SF,on =['Flow id','Time'])[['Flow id','Length']].unique()
FPL = FPL.groupby(['Flow id'],{
        'FPL':turicreate.aggregate.AVG('Length')
    })

SF = SF.join(FPL,on='Flow id')
SF.head()
```

```
## Number of packets per flow
temp = SF.groupby(['Flow id'],{
        'NumPackets':turicreate.aggregate.COUNT()
    })

SF = SF.join(temp, on ='Flow id')
del(temp)
```

```
SF.head()
```

```
## Number of bytes exchanged
temp = SF.groupby(['Flow id'],{
        'BytesTotal':turicreate.aggregate.SUM('Length')
    })
SF = SF.join(temp, on ='Flow id')
del(temp)
```

```
SF.head()
```

```
## Standard deviation of packet length
temp = SF.groupby(['Flow id'],{
        'StdDevLen':turicreate.aggregate.STDV('Length')
    })
SF = SF.join(temp, on ='Flow id')
del(temp)
```

```
SF.head()
```

```
# ## Same length packet ratio
# temp2 = SF.groupby(['Flow'],{
#         'SameLenPktRatio':turicreate.aggregate.COUNT_DISTINCT('Length')
#     })
# ##temp from number of packets computation
# temp = SF.groupby(['Flow'],{
#         'NumPackets':turicreate.aggregate.COUNT()
#     })
# temp = temp.join(temp2,on='Flow')
# temp['SameLenPktRatio'] = temp['SameLenPktRatio']*1.0/temp['NumPackets']
# temp2 = None
# temp = temp[['Flow','SameLenPktRatio']]
# SF = SF.join(temp, on ='Flow')
```

```
temp = SF.groupby(['Flow id'],{
        'SameLenPktRatio':turicreate.aggregate.COUNT_DISTINCT('Length'),
        'NumPackets':turicreate.aggregate.COUNT()
    })
temp['SameLenPktRatio'] = temp['SameLenPktRatio']*1.0/temp['NumPackets']
temp = temp[['Flow id','SameLenPktRatio']]
SF = SF.join(temp, on ='Flow id')
```

```
SF.head()
```

```
## Duration of flow
timeF = SF.groupby(['Flow id'],{
        'startTime':turicreate.aggregate.MIN('Time'),
        'endTime':turicreate.aggregate.MAX('Time')
    })
timeF['Duration'] = timeF['endTime'] - timeF['startTime']
timeF = timeF[['Flow id','Duration']]
SF = SF.join(timeF, on ='Flow id')
```

```
SF.head()
```

```
## Average packets per second
temp =  SF.groupby(['Flow id'],{
        'NumPackets':turicreate.aggregate.COUNT()
    })
temp = temp.join(timeF,on=['Flow id'])
temp['AvgPktPerSec'] = temp.apply(lambda x:0.0 if x['Duration'] == 0.0 else x['NumPackets']*1.0/x['Duration'])
temp = temp[['Flow id','AvgPktPerSec']]
SF = SF.join(temp, on ='Flow id')
```

```
SF.head()
```

```
##Average Bits Per Second
temp = SF.groupby(['Flow id'],{
        'BytesEx':turicreate.aggregate.SUM('Length')
    })
temp = temp.join(timeF,on=['Flow id'])
temp['BitsPerSec'] = temp.apply(lambda x:0.0 if x['Duration'] == 0.0 else x['BytesEx']*8.0/x['Duration'])
temp = temp[['Flow id','BitsPerSec']]
SF = SF.join(temp, on ='Flow id')
```

```
SF.head()
```

```
## Average Packet Lentgth
temp = SF.groupby(['Flow id'],{
        'APL':turicreate.aggregate.AVG('Length')
    })
SF = SF.join(temp, on ='Flow id')
```

```
SF.head()
```

```
## Inter arrival time
SF['IAT'] = 0
# SF = SF.sort(['FlowNo.','Time'])
prev = None
prevT = None
li = []
for x in SF:
    if prev is None or x['Flow id']!= prev:
        li.append(0)
    else:
        li.append(x['Time']-prevT)        
    prev = x['Flow id']
    prevT = x['Time']
SF['IAT'] = turicreate.SArray(li)
```

```
SF.head()
```

```
## Null Packets handling
def checkNull(x):
    if(x['TCP Segment Len']=='0' or x['udp_Length']==8 ):
        return 1
    elif('ipx' in x['Protocols in frame'].split(':')):
        l = x['Length'] - 30
        if('eth' in x['Protocols in frame'].split(':')):
            l = l - 14
        if('ethtype' in x['Protocols in frame'].split(':')):
            l = l - 2
        if('llc' in x['Protocols in frame'].split(':')):
            l = l - 8
        if(l==0 or l==-1):
            return 1
    return 0
```

```
SF['isNull'] = SF.apply(lambda x:checkNull(x))
```

```
SF.head()
```

```
NPEx = SF.groupby(['Flow id'],{
        'NPEx':turicreate.aggregate.SUM('isNull')
    })
SF = SF.join(NPEx, on ='Flow id')
```

```
SF.head()
```

```
# ### Number of Reconnects - considering only TCP reconnects, using sequence number
recon = SF[SF['Sequence number']!=''].groupby(['Flow id'],{
        'total_seq_no.' : turicreate.aggregate.COUNT('Sequence number'),
        'distinct_seq_no.' : turicreate.aggregate.COUNT_DISTINCT('Sequence number')
    })
recon['reconnects'] = recon['total_seq_no.'] - recon['distinct_seq_no.']
recon.head()
recon = recon[['Flow id','reconnects']]
SF = SF.join(recon,on='Flow id',how='left')
len(SF)
```

```
SF.head()
```

```
#To identify records where reconnect check was not applied like UDP etc
SF.fillna('reconnects',-1)
```
